## Import modules

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

%matplotlib inline
%pylab inline
```

One need to produce the dataset.

```python
circle_x = 5.                          # para fijar el centro del circulo
circle_y = 5.                          #
L=0.25
N=1000
phi = 2. * np.pi * np.random.rand(N)

ranillo = L*np.random.uniform(0.,1.,N)
rani = np.sqrt(ranillo)
x1 = rani * np.cos(phi) #+ circle_x
y1 = rani * np.sin(phi) #+ circle_y

rani2 = np.sqrt(4*L*np.random.uniform(1.,1.5,N))
x2 = rani2 * np.cos(phi)
y2 = rani2 * np.sin(phi)

plt.plot(x1,y1,'.', label = 'hobbits')
plt.plot(x2,y2,'.', label = 'orcos')
plt.xlabel('x',fontsize = 20)
plt.ylabel('y',fontsize = 20)
plt.axis('equal')
plt.legend(fontsize='x-large')
```

```python
x = np.vstack((x1,x2)) # shape (2,1000)
y = np.vstack((y1,y2)) #
```
